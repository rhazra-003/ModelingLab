```python
import numpy as np
```

```python
def get_min_min2(array):
    m1, m2 = np.inf,np.inf
    
    for val in array:
        if val <= m1:
            m1, m2 = val, m1
        elif val <= m2:
            m2 = val
    
    return m1, m2

def get_max_penalty(cost, n_warehouse, n_store):
    # for row
    row_penalty = []
    
    for r in range(n_warehouse):
        m1, m2 = get_min_min2(cost[r])
        
        if m1 != np.inf and m2 != np.inf:
            row_penalty.append((m2-m1, r))
        elif m1 != np.inf:
            row_penalty.append((m1, r))
    
    # for coloumn
    col_penalty = []
    
    for c in range(n_store):
        col_m = []
        for i in range(n_warehouse):
            col_m.append(cost[i][c])
        m1, m2 = get_min_min2(col_m)
        
        if m1 != np.inf and m2 != np.inf:
            col_penalty.append((m2-m1, c))
        elif m1 != np.inf:
            col_penalty.append((m1, c))
            
    print("Row penalty (p, row):", row_penalty)
    
    print("Column penalty (p, col):", col_penalty)
    
    if not (row_penalty or col_penalty):
        return -1, -1, None
    
    r_mp, c_mp = max(row_penalty), max(col_penalty)
    
    if r_mp > c_mp:
        i = r_mp[1]
        mp = r_mp[0]
        j=0
        mn = cost[i][j]
        
        for c in range(n_store):
            if cost[i][c] < mn:
                j = c
                mn = cost[i][c]
                
    else:
        j = c_mp[1]
        mp = c_mp[0]
        i=0
        mn = cost[i][j]
        
        for r in range(n_warehouse):
            if cost[r][j] < mn:
                i = r
                mn = cost[i][c]
    
    return i, j, mp

supply = []
demand = []
cost = []
sum_demand, sum_supply, total_cost, no_alloc = 0, 0, 0, 0

n_warehouse = int(input("Enter the number of Warehouses: "))
n_store = int(input("Enter the number of Stores: "))

supply = list(map(int,input("Enter the supply matrix: ").strip().split()))[:n_warehouse]
demand = list(map(int,input("Enter the demand matrix: ").strip().split()))[:n_store]

print("Enter the cost matrix: ")

for i in range(n_warehouse):
    r = list(map(int, input().strip().split()))[:n_store]
    cost.append(r)

sum_supply = sum(supply)
sum_demand = sum(demand)

if(sum_supply == sum_demand):
    print("\nGiven Transportation Problem is Balanced!\n")
else:
    print("\nGiven Transportation Problem is Not Balanced!\n")

solution = []

for k in range(n_warehouse):
    temp = [0]*n_store
    solution.append(temp)
    
i, j, max_penalty = get_max_penalty(cost, n_warehouse, n_store)

while(max_penalty != None):
    print("selected for allocation, i=",i," j=",j)
    
    x = min(supply[i], demand[j])
    
    supply[i] = supply[i] - x
    demand[j] = demand[j] - x
    
    no_alloc = no_alloc + 1  
    total_cost = total_cost + x*cost[i][j]
    
    if x > 0 :
        solution[i][j] = x
    
    if supply[i] == 0:
        for j in range(n_store):
            cost[i][j] = np.inf
    
    if demand[j] == 0:
        for i in range (n_warehouse):
            cost[i][j] = np.inf

    i, j, max_penalty = get_max_penalty(cost, n_warehouse, n_store)
    
print("\nTransportation Distribution: ")

for i in range(n_warehouse):
    for j in range(n_store):
        if solution[i][j] > 0:
            print(solution[i][j],"units from Plant ",i+1," to Warehouse ",j+1)    

print("\nTotal cost of transportation: ", total_cost)

if no_alloc != n_warehouse + n_store - 1:
    print("\nThis is a Degenerate Solution! \n")
    
else:
    print("\nThis is a Non-Degenerate Solution!\n")
```
