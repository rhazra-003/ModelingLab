```python
C = [[2, 3, 11, 7],
        [1, 0, 6, 1],
        [6, 8, 15, 9],
        [7, 5, 3, 2]]

S=[6,1,10]
D=[7,5,3,2]

T_cost = 0
No_alloc = 0

supply=[]

i = 0
j = 0
m = 4
n = 3
```

```python
if(np.sum(S)!=np.sum(D)):
    sys.exit("Unbalanced")
```

```python
while i<n and j<m:
    x = min(S[i], D[j])
    S[i] -= x
    D[j] -= x
    No_alloc += 1
    if x>0:
        supply.append([x, i+1, j+1])
    T_cost += C[i][j]*x
    if S[i] == 0:
        i += 1
    if D[i] == 0:
        j += 1

if No_Alloc != m+n-2:
    sys.exit("Degeneracy condition not fulfilled")

print("Total cost is: ", T_cost)
    
for it in supply:
    print(it[0], "units sent from warehouse", it[1], "to store", it[2]) 
```

```python
C = [[9, 12, 9, 6, 9, 10],
             [7, 3, 7, 7, 5, 5],
             [6, 5, 9, 11, 3, 11],
             [6, 8, 11, 2, 2, 10]]

T_Cost = 0
No_Alloc = 0
i = 0
j = 0
m = 6
n = 4

S = [5, 6, 2, 9]
D = [4, 4, 6, 2, 4, 2]
```

```python
if(np.sum(S)!=np.sum(D)):
    sys.exit("Unbalanced")
```

```python
supply = []
while True:
    minCost = np.amin(C)
    if minCost == np.amin(C):
        if minCost == np.inf:
            break
        result = [np.inf, -1, -1]
        for i in range(n):
            for j in range(m):
                if C[i][j] == minCost:
                    if result[1] == -1 or min(S[i], D[j]) > min(S[result[1]], D[result[2]]):
                        result = [C[i][j], i, j]
    
    x = min(S[result[1]], D[result[2]])
    No_Alloc += 1
    T_Cost += result[0]*x
    S[result[1]] -= x
    D[result[2]] -= x
    
    supply.append([x, result[1] +1, result[2] +1])
    if S[result[1]] == 0:
        for j in range(m):
            C[result[1]][j] = np.inf
    if D[result[2]] == 0:
        for i in range(n):
            C[i][result[2]] = np.inf
            
print(No_Alloc)
    
if No_Alloc != m+n-2:
    sys.exit("Degeneracy condition not fulfilled")
print("Total cost is: ", T_Cost)
    
for it in supply:
    print(it[0], "units sent from warehouse", it[1], "to store", it[2])
```

```python
import numpy as np

noAlloc = 0
inf = np.iinfo(np.int32).max

r = int(input("Enter the number of rows:"))
c = int(input("\nEnter the number of columns:"))

print("\nEnter the entries in a single line (separated by space):\n")
entries = list(map(int, input().split()))
costMatrix = np.array(entries).reshape(r, c)
print("\nCost Matrix")
print(costMatrix)

print("\nRows, Columns: (",r,",",c,")")
print("\nEnter Supply values: ")
supplyMatrix = list(map(int, input().split()))
print("\nSupply Matrix: ")
print(supplyMatrix)

print("\nEnter Demand values: ")
demandMatrix = list(map(int, input().split()))
print("\nDemand Matrix: ",sep = '')
print(demandMatrix)

total_demand = np.sum(demandMatrix)
total_supply = np.sum(supplyMatrix)

if total_demand == total_supply:
    print("\nBalanced Transportation Problem.")
else:
    print("\nUnbalanced Transportation Problem")

if total_demand < total_supply:
    costMatrix = np.insert(costMatrix, c-1, 0, axis = 1)
    costMatrix[r-1][c-1] = total_supply - total_demand
    c += 1
else:
    costMatrix = np.insert(costMatrix, [r-1], [[0],], axis = 0)
    costMatrix[r-1][c-1] = total_demand - total_supply
    r += 1
   
print()
i=0
j=0
A = np.zeros([r, c], dtype=int)
min_cost = np.amin(costMatrix)

while min_cost != inf:
    indices = np.where(costMatrix == min_cost)
    i = indices[0][0]
    j = indices[1][0]
    x = min(supplyMatrix[i],demandMatrix[j])
    supplyMatrix[i] -= x
    demandMatrix[j] -= x
    noAlloc += 1
    A[i][j] = x

    if supplyMatrix[i] < demandMatrix[j]:
        j = 0
        while j < c:
            costMatrix[i][j] = inf
            j += 1
    elif supplyMatrix[i] > demandMatrix[j]:
        i = 0
        while i < r:
            costMatrix[i][j] = inf
            i += 1
    else:
        k = 0
        while k < c:
            costMatrix[i][k] = inf
            k += 1
        k = 0
        while k < r:
            costMatrix[k][j] = inf
            k += 1
        min_cost = np.amin(costMatrix)

print("\nNo of Allocation: ", noAlloc)
print("\nAllocation Matrix:\n")
print(A)

if (r + c - 41) == noAlloc and total_demand == total_supply:
    print("\nIt is not a Degenerate Solution")
else:
    print("\nDegenerate Solution")
```
